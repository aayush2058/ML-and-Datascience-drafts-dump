```python
# reading different dataframe
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
#creating a histogram of age
heart_disease.age.plot.hist(bins = 50)
```

```python
heart_disease.plot.bar(subplots = True, figsize = (10,50))
```

```python
over_50 = heart_disease[heart_disease.age > 50]
over_50
```

```python
# Pyplot method
over_50.plot(figsize = (10, 6),
             kind = "scatter",
            x = 'age',
            y = 'chol',
            c = 'target')
```

```python
# OO method mixed with pyplot
fig, ax = plt.subplots(figsize = (10 , 6))
over_50.plot(kind = 'scatter',
            x = 'age',
            y = 'chol',
            c = 'target',
            ax = ax)
ax.set_xlim(45, 80)
```

## OO method from scratch

```python
fig, ax = plt.subplots(figsize = (10, 6))

scatter = ax.scatter(x = over_50.age,
                     y = over_50.chol,
                     c = over_50.target)

ax.set(xlabel = "age",
            ylabel = 'cholesterol',
            title = "age with cholesterol level")

ax.legend(*scatter.legend_elements(), title = "Title");

ax.axhline(over_50.chol.mean(),
           linestyle = '--' );
```

# Matplotlib in a nutshell

```python
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (5, 5),
                              sharex = True)

scatter = ax0.scatter(x = over_50.age,
                     y = over_50.chol,
                     c = over_50.target,
                     cmap = 'summer')
ax0.set(title = " age with cholesterol level ",
        ylabel = "cholesterol level")

ax0.legend(*scatter.legend_elements(), title = "target")

ax0.axhline(over_50.chol.mean(), linestyle = '--')

ax0.set_xlim([50, 80])

## for heart rate we are taking ax1

scatter = ax1.scatter(x = over_50.age,
                      y = over_50.thalach,
                      c = over_50.target,
                     cmap = 'winter')

ax1.set(title = "age with heart rate",
       xlabel = "age",
       ylabel = "heart-rate / thalach")

ax1.legend(*scatter.legend_elements(), title = "target")

ax1.axhline(over_50.thalach.mean(), linestyle = (0, (3, 10, 1, 10, 1, 10)));

fig.suptitle("Heart disease analysis", fontsize = 16, fontweight = "bold");
```
